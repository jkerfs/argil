```python
from argil.visualization import d3
import argil.environment
from argil.entity import Agent
```

```python
width = 500
height = 300


xpos =  np.arange(0, width, 1)
ypos = 100 * np.sin( xpos / 20) + 100
params = {"xpos": xpos, "ypos": ypos, "ind": 0}

def step(self, get_obstacles, get_neighbors):
    self.ind += 1
    self.ind %= self.xpos.size
    self.x = self.xpos[self.ind]
    self.y = self.ypos[self.ind]

def glance(self):
    return {"_shape": "circle", "r": 10, "fill": "blue"}

def observe(self):
    return {"cx": self.x, "cy": self.y}
```

```python
agent = Agent(step, **params)
param2 = {"xpos": xpos + 30, "ypos": ypos, "ind": 0, "_line": ["x","y"]}
agent2 = Agent(step, **param2)
roboWorld = argil.environment.Environment([agent,agent2], [], width, height)
```

```python
simulation = d3.D3Simulation(roboWorld, observe, glance)
```

```python
simulation.run(num_steps=500, speed=1)
```
